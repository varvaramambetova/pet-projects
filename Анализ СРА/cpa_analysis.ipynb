### Исследовательский анализ данных

#### CPA общий по месяцам 

```python
cpa_month = data_new.pivot_table(index =['Month'], values = ["Cost", "Conversions"], aggfunc = 'sum')
cpa_month['cpa_month'] = cpa_month['Cost']/cpa_month['Conversions']
cpa_month
```

```python
cpa_month.pivot_table(index =cpa_month.index, values = 'cpa_month').plot(grid=True, figsize=(8, 3))
plt.ylabel('CPA')
plt.xlabel('Месяц привлечения ')
plt.title('CPA общий по месяцам')
plt.show()
```

#### CTR % по месяцам

```python
ctr_month = data_new.pivot_table(index =['Month'], values = ['Shows', "Conversions", "Clicks","Cost"], aggfunc = 'sum')
ctr_month['ctr_month,%'] = ctr_month['Clicks']/ctr_month['Shows']*100
ctr_month['cr_month,%'] = ctr_month['Conversions']/ctr_month['Clicks']*100
ctr_month
```

```python
ctr_month.pivot_table(index =ctr_month.index, values = 'ctr_month,%').plot(grid=True, figsize=(8, 3))
plt.ylabel('CTR, %')
plt.xlabel('Месяц привлечения ')
plt.title('CTR % общий по месяцам')
plt.show()
```

#### CR % по месяцам

```python
ctr_month.pivot_table(index =ctr_month.index, values = 'cr_month,%').plot(grid=True, figsize=(8, 3))
plt.ylabel('CR, %')
plt.xlabel('Месяц привлечения ')
plt.title('CR % общий по месяцам')
plt.show()
```

#### Расходы по месяцам

```python
ctr_month.pivot_table(index =ctr_month.index, values = 'Cost').plot(grid=True, figsize=(8, 3))
plt.ylabel('Расходы')
plt.xlabel('Месяц привлечения ')
plt.title('Расход общий по месяцам')
plt.show()
```

- СРА за 4 последних месяца не соответствовал целевому значению.
- За последние 4 месяца отмечалась динамика увеличения СРА, при этом, кликабельность и конверсия пользователей снижались.
- С июня по июль равномерное повышение СРА, кликабельности и конверсии,при этом также отмечался рост расходов на рекламу, все показатели росли равномерно,СРА повышался за счет увеличения расходов.
- Затем с июля по август произошел резкий спад кликабельности и процент целевых действий,что привело к повышению СРА.
- С августа по сентябрь видимо решили, что выравнивания СРА нужно снизить расходы на рекламу, кликабельность и процент конверсий упали, а СРА поднялся еще выше. 
- Видим следующую картину: расходы снизили ,кликабельность, конверсия снизились и СРА вырос. 
Среди возможных причин повышения СРА при снижении расходов можно выделить :

- Динамика СРА по типу Retargeting больше в несколько раз целевого значения и имеет тенденцию к повышению, кликабельность и конверсия по данному типу таргетинга самая низкая среди всех, по всей видимости и цена за целевое действие самая высокая. При этом кликабельность и конверсия в динамике снижаются, что на фоне увеличения расходов привело к тому, что СРА резко выросло. 

 Причины повышения СРА, снижения конверсии и кликабельности ретаргетинга:
- Утомление аудитории: Повторный показ рекламы одной и той же продукции или услуги одной и той же аудитории мог привести к утомлению, что снизило их желание кликнуть на объявление и совершить целевое действие.
- Неверный таргетинг: Неправильная настройка таргетинга могла привести к тому, что ретаргетинговая реклама не попала к нужной аудитории, что снизило ее эффективность.

- Качество объявлений: Если объявления ретаргетинга были не привлекательны или не соответствовали ожиданиям аудитории, это могло снизить их кликабельность и подвергнуть конверсию риску.

 Рекомендации по снижению СРА до целевого значения по таргетингу:
- Для автотаргетинга настройка анализ  целевой страницы, анализ страниц на призыв к действию , анализ технических аспектов отображения  целевой страницы, улучшение продукта.
Для фразового изменение настроек таргетинга,   технических аспектов отображения рекламных баннеров, 
- Сегментация аудитории: для создания более релевантных и персонализированных объявлений для каждой группы.
- A/B-тестирование различных изображений, заголовков и предложений в ретаргетинговых объявлениях.
- Постепенная частота показа ретаргетинговых объявлений, чтобы избежать назойливости.
- Оптимизация лендингов: оптимизация для конверсий,связаны с объявлениями и предлагать ясный и привлекательный call-to-action.

- Плохая оптимизация баннеров для мобильных устройств
- Нецелевой трафик: плохо настроенный таргетинг
- Неудовлетворительная целевая страница для мобильных устройств
- Отсутствие ясного call-to-action: Даже если баннер  привлекал клики, отсутствие четкого призыва к действию на целевой странице могло уменьшить конверсию.
- Содержание целевой страницы не соответствовало ожиданиям, сформированным после просмотра баннера.
3.  Возможные причины роста СРА в категории РКО на мобильных устройствах по Retargeting: 
- Нецелевой трафик: Если ретаргетинг  был настроен на аудиторию, которая не проявляла реального интереса к продукту или услуге.
-  Утомленность аудитории: Пользователи, на которых направлен ретаргетинг, могли быть утомлены от постоянного видения одной и той же рекламы на мобильном устройстве.
- Рекламный контент при ретаргетинге мог не вызывать достаточного интереса или не соответствовать ожиданиям целевой аудитории.
- Возможно ретаргетинговые объявления и целевая страница были не оптимизированы для мобильных устройств.
4. Возможные причины роста СРА По баннеру типа text категории РКО площадки Networks:
- Если текстовый контент баннера был не привлекателен это могло привести к понижению CTR (кликабельности) и росту CPA.
- Несоответствие контексту: Баннер не соответствовал контексту, в котором он отображался. 
5. Возможные причины роста СРА ретаргетинга категории РКО на площадке Networks: 
- Утомление аудитории: Повторные показы через ретаргетинг могли привести к утомлению аудитории
- Нецелевой таргетинг: Возможно, реклама показывалась аудитории, которая уже не заинтересована.
- Качество креатива: Если объявления ретаргетинга не привлекательны или не соответствовали ожиданиям аудитории, это могло увеличить кликабельность, но снизить конверсию из-за недостаточной привлекательности предложения.


### Рекомендации по снижению СРА

 Рекомендации для снижения СРА по устройствам : 
- Провести тестирование разных типов баннеров для Desktop
- провести более тщательный анализ таргетинга на Desktop устройствах
- провести анализ целевой страницы при переходе с мобильных устройств: технические аспекты отображения
- создать креативы целевой страницы, которые бы мотивировали аудиторию мобильных устройств к действию.
- Осуществлять мониторинг и анализ результатов, чтобы быстро реагировать на изменения в кликабельности и стоимости привлечения клиента.
- Уделять внимание техническим аспектам, чтобы устранить возможные проблемы с отображением рекламных баннеров на устройствах Desktop.

 Рекомендации по снижению СРА на площадке Networks :
 
- Проводить A/B-тестирование различных вариантов лендингов и рекламных материалов для выявления наиболее эффективных решений.
- Тщательно настроить таргетинг на площадке: Убедиться, что рекламные кампании нацелены на аудиторию с высоким потенциалом конверсии.
- Оптимизировать рекламные материалы: Разработать привлекательные и релевантные объявления, которые будут привлекать внимание посетителей Networks.

 Рекомендации по снижению СРА по типу баннера :
 
- На время медийную рекламу приостановить
- Повышение расходов на текстовой баннер для повышения трафика
- Оптимизировать дизайн и содержание баннеров: Создавать привлекательные и информативные изображения с четким сообщением, которые будут мотивировать аудиторию к действию.
- Тщательно настраивать таргетинг по баннеру.
- Тестировать различные варианты баннеров image_ad  и настроек кампаний, а затем анализировать результаты для определения наиболее эффективных стратегий.  

 Рекомендации по снижению СРА до целевого значения по таргетингу:
 
- Для автотаргетинга настройка анализ  целевой страницы, анализ страниц на призыв к действию , анализ технических аспектов отображения  целевой страницы, улучшение продукта.
- Для фразового изменение настроек таргетинга,   технических аспектов отображения рекламных баннеров, 
- Сегментация аудитории: для создания более релевантных и персонализированных объявлений для каждой группы.
- A/B-тестирование различных изображений, заголовков и предложений в ретаргетинговых объявлениях.
- Постепенная частота показа ретаргетинговых объявлений, чтобы избежать назойливости.
- Оптимизация лендингов: оптимизация для конверсий,связаны с объявлениями и предлагать ясный и привлекательный call-to-action.

 Рекомендации по снижению СРА по типу запроса :

- Оптимизация ключевых слов: Провести пересмотр и оптимизацию списка ключевых слов для запросов Other и Competitor, включая более точные и релевантные фразы, что может помочь улучшить таргетинг и снизить стоимость привлечения.
- Улучшение объявлений: Создать более привлекательные и релевантные объявления, которые будут лучше привлекать внимание аудитории и побуждать их к действию, таким образом повышая кликабельность.
- Тестирование и анализ: Проводить регулярное тестирование изменений в кампаниях, а также анализировать результаты для определения наиболее эффективных стратегий и тактик.
- Оптимизация лендинга: Обеспечить, чтобы страницы назначения (лендинги) были оптимизированы для конверсий, убедившись в их соответствии ожиданиям аудитории и удобстве использования.
